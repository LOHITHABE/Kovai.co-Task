```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_csv("Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20250603.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()
```

```python
df = df.sort_index()
df.head()
```

```python
df = df.asfreq('D')
```

```python
services = df.select_dtypes(include=['number']).columns.tolist()
services
```

```python
def sarima_forecast(series, steps=7):
    model = SARIMAX(series,
                    order=(1,1,1),
                    seasonal_order=(1,1,1,7),
                    enforce_stationarity=False,
                    enforce_invertibility=False)
    model_fit = model.fit(disp=False)
    forecast = model_fit.forecast(steps=steps)

    forecast = forecast.round()

    forecast = forecast.clip(lower=0)
    
    forecast = forecast.astype(int)

    return forecast
```

```python
forecast_results = {}
for service in services:
    forecast_results[service] = sarima_forecast(df[service], 7)
forecast_results
```

```python
forecast_df = pd.DataFrame(forecast_results)
forecast_df
```

```python
forecast_df.plot(figsize=(8,4))
plt.title("7-Day Passenger Forecast for All Transport Services")
plt.xlabel("Days Ahead")
plt.ylabel("Forecasted Journeys")
plt.show()
```
